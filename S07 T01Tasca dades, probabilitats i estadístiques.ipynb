## Exercise 1 Mode, median, standard deviation and aritmetic mean

```python
# Finding the mode of column Speed

mode= statistics.mode(df["SPD"])
print ("The Speed mode, the most frequent observation, is: " + str(mode))
```

```python
# Finding the median of column Speed

median=df["SPD"].median()
print ("The Speed median, the value that lies in the middle when we sort the data, is: "+ str(median))
```

```python
# Finding the standartd deviation of column Speed

sd=df["SPD"].std()
print ("The Speed standard deviation, a measure of how spread out numbers are, is: " + str (sd))
```

```python
# Finding the mean of column Speed

mean=df["SPD"].mean()
print ("The Speed mean, the average of the data, is: " + str (mean))
```

```python
sns.histplot(df["SPD"], kde=True)
plt.title('Speed histogram',fontsize=15)
```

The mode and the median have the same value, but the average is lower. So I have plotted an histogram to see more easily the distribution, which looks like a bivariate distribution because in many sports speed is no so important as other skills, for example, in curling, fishing or golf.

## Exercise 3 Calculate all the atribute's correlations¬∂

```python
corr = df.iloc[:,:-1].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(10, 8))
    ax = sns.heatmap(corr, mask=mask, annot=True, vmax=1, vmin=-1, square=True)
    
plt.title('Correlation between all the atributes',fontsize=15)
plt.show()
```

The skills with higher correlation are Power and Strength, which are very close, analytical aptitude and hand-eye cordination and agility and speed. The most correlated with Total are Durability and Agility, followed by Speed, which could be the skills that add most to the final result.

## Exercise 4 Calculate geometric mean and harmonic mean.

```python
from scipy.stats.mstats import gmean
```

```python
# Finding the geometric mean of Durability.

geometricMean = gmean(df.loc[:,'DUR'])

print ('The Geometric Mean of Durability is: ' + str(geometricMean)) 
```

The geometric mean is a series of numbers calculated by taking the product of these numbers and raising it to the inverse of the length of the series.Formula for geometric mean is {[(1+Return1) x (1+Return2) x (1+Return3)‚Ä¶)]^(1/n)]} ‚Äì 1 and for arithmetic mean is (Return1 + Return2 + Return3 + Return4)/ 4.
Geometric mean can only be calculated for positive numbers and is always less than geometric meanwhile, arithmetic mean can be calculated for both positive and negative numbers and is always greater than the geometric mean. A most common problem with having a dataset is the effect of outliers. Geometric mean normalizes the dataset, and the values are averaged out. The geometric mean is useful in cases where the dataset is logarithmic or varies by multiples of 10.
For our dataset, whichs hasn't outlier nor is logarithmic, is of no use.

```python
# Finding the harmonic mean of Durability.

harmonicMean = statistics.harmonic_mean(df.DUR)
print ('The Harmonic Mean is: ' + str(harmonicMean)) 
```

Harmonic mean is a type of average that is calculated by dividing the number of values in a data series by the sum of the reciprocals (1/x_i) of each value in the data series. The harmonic mean is often used to calculate the average of the ratios or rates. It is the most appropriate measure for ratios and rates because it equalizes the weights of each data point. For instance, the arithmetic mean places a high weight on large data points, while the geometric mean gives a lower weight to the smaller data points. For our dataset, with no ratios or outliers, is of no use.
